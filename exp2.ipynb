```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import pandas as pd,numpy as np

dataset=load_iris()
x=pd.DataFrame(dataset.data)
#x.columns=['Sepal_length','Sepal_width','Petal_length','Petal_width']
x
```

```python
y=dataset.target
```

```python
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
scaler.fit(x)
```

```python
xsa=scaler.transform(x)
xs=pd.DataFrame(xsa,columns=x.columns)
```

```python
gmm=GaussianMixture(n_components=3)
kmeans=KMeans(n_clusters=4)
```

```python
gmm.fit(xs)
kmeans.fit(xs)
```

```python
y_cluster_gmm=gmm.predict(xsa)
k_cluster_kmeans=kmeans.predict(xsa)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy for GMM= ", accuracy_score(y,y_cluster_gmm))
print("Accuracy for Kmeans= ", accuracy_score(y,k_cluster_kmeans))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
colormap=np.array(['red','lime','black','yellow'])
plt.subplot(1,2,1)
plt.scatter(x.Petal_length,x.Petal_width,c=colormap[y],s=40)
plt.title('Real')
```

```python
plt.subplot(1,2,2)
plt.scatter(x.Petal_length,x.Petal_width,c=colormap[y_cluster_gmm],s=40)
plt.title("GMM clustering")
```

```python
plt.subplot(1,3,3)
plt.scatter(x.Petal_length,x.Petal_width,c=colormap[k_cluster_kmeans],s=40)
plt.title("Kmeans clustering")
```
